```python
import tensorflow
# from tensorflow.one_hot import one_hot
```

```python
## sampel sentences
sent=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'i am a good boy',
    'i am  a good developer',
    'understand the meaning of words',
    'your videos are good',
]
sent
```

```python
## define the vocabulary size
voc_size=10000
```

```python
## one hot representation
from tensorflow.keras.preprocessing.text import one_hot
one_hot_repr=[one_hot(words, voc_size) for words in sent]
one_hot_repr
```

```python
## word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
# from tf.keras.preprocessing import pad_sequence   ===> in older versions
# from tensorflow.keras.utils import pad_sequence
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
import numpy as np
```

```python
sent_length=8
embedded_doc=pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
embedded_doc
```

```python
## feature representation
dim=10
```

```python
model = Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=dim, input_shape=(sent_length,)))
model.compile('adam', 'mse')
model.summary()
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
